```python
# Importing the packages

# Data cleaning and manipulation 
import pandas as pd 
import numpy as np

# Iterating values
import itertools

# Data visualization
import matplotlib.pyplot as plt

# Graph making
import networkx as nx

# To warn on situations that arent exceptions
import warnings
```

```python
# Reading the data from Course.csv and creating a DataFrame with it.
data = pd.read_csv("Course.csv", delimiter = ',') 

# DataFrame isÂ a 2-dimensional labeled data structure with columns of potentially different types. 
# You can think of it like a spreadsheet or SQL table, or a dict of Series objects. It is generally the most commonly 
# used pandas object.
df = pd.DataFrame(data) 

# Printing out our values
df
```

```python
# Finding the number of courses and semesters and printing it.

# The unique() function is used to find the unique elements of an array. 
# Returns the sorted unique elements of an array.
courses = df['Course'].unique()
n = len(courses)
print(f"Courses chosen: {courses}")
print(f"Distinct number of courses: {n}")

sem = df['Sem'].unique()
s = len(sem)
print(f"\nSemesters: {sem}")
print(f"Distinct number of semesters: {s}")
```

```python
# Creates an empty dictionary called subjects which will contain all possible combinations of subjects for each semester
subjects = {}
i = 0

# The code starts by creating a list of the subjects that are being analyzed.
# It then loops through each subject and creates a new list with the number of students in that subject.

for s1, s2, s3, s4, s5, s6, c in zip(df['Sub1'], df['Sub2'], df['Sub3'], df['Sub4'], df['Sub5'], df['Sub6'], df['Course']):
    
    # The code then iterates through each student and checks if they're enrolled in their own course or not. 
    # If they are enrolled in their own course, then it increments i by 1 and keeps going until it finds a 
    # match for s1 with another student's name
    # The code iterates through the list of subjects and updates the list with each new subject, incrementing the index by one.
    if s1 not in subjects and s1 == s1:
        subjects.update({s1:i})
        i = i + 1
    elif s2 not in subjects and s2 == s2:
        subjects.update({s2:i})
        i = i + 1
    elif s3 not in subjects and s3 == s3:
        subjects.update({s3:i})
        i = i + 1
    elif s4 not in subjects and s4 == s4:
        subjects.update({s4:i})
        i = i + 1
    elif s5 not in subjects and s5 == s5:
        subjects.update({s5:i})
        i = i + 1
    elif s6 not in subjects and s6 == s6:
        subjects.update({s6:i})
        i = i + 1
        
print("\nTotal Number of subject is:", len(subjects))

# It starts by checking if there is an entry for s1 in subjects, which would mean that this student has already been 
# assigned to one course 
# If there is no entry for s1, then it will assign them to Course 1 and increment i by 1.
# Otherwise, if there is an entry for s2 or s3 or so on, then it will assign them to Course 2 or 3 respectively and 
# increment i by 1 again.
# This continues until all entries have been accounted for and i values
# The code iterates through the list of subjects, and checks if the subject exists in the list.
# If it does not exist, then it is added to the list.
```

```python
# To translate through all subjects and then through all the rows
matrix = [[0 for i in range(len(subjects))] for j in range(s*n)]

# The code starts by initializing a matrix with the number of rows equal to the number of subjects
for s1, s2, s3, s4, s5, s6, sem, course in zip(df['Sub1'], df['Sub2'], df['Sub3'], df['Sub4'], df['Sub5'], 
                                               df['Sub6'], df['Sem'], df['Course']):
    
    # The code then iterates through each subject and assigns 1 to that subject if it is in semester 1, 2 or 3
    if course == 'Mtech':
        # if CSE, it would go to the other if and cuz its Mtech, it has to be sem + 3 as first 3 are CSE
        # So next 3 should be Mtech
        sem = sem + 3
        
    # Next it checks if there is a value for that subject in the matrix array. 
    # If there is then it increments that element by one.
    if s1 == s1:
        matrix[sem-1][subjects[s1]] = 1
    if s2 == s2:
        matrix[sem-1][subjects[s2]] = 1
    if s3 == s3:
        matrix[sem-1][subjects[s3]] = 1
    if s4 == s4:
        matrix[sem-1][subjects[s4]] = 1
    if s5 == s5:
        matrix[sem-1][subjects[s5]] = 1
    if s6 == s6:
        matrix[sem-1][subjects[s6]] = 1

# The code is a for loop that iterates through the number of subjects in the list. 
# The code then assigns a value to each subject based on the iteration of the loop. 
# The code above initializes a matrix with 0s for all positions, which will be used as an index into 
# the dataframe to assign values to each subject.        

# Printing the frame for the matrix
Student_matrix = pd.DataFrame(matrix, columns = subjects.keys(), index = ['CSE', 'CSE', 'CSE', 'Mtech', 'Mtech', 'Mtech'])
print("\nList of Subject semester wise:")
Student_matrix
```

```python
# Returning a dictionary with the keys of subjects and values of the students.
reverse = dict(zip(subjects.values(),subjects.keys()))
reverse
```

```python
# The code starts by creating a nx.Graph object called G, which is the complete graph for the given subject of sem 1.
# The code then iterates through each sem and creates an empty list called sub 
# that contains all possible colors in order from 0 to 6. 

# colour for each subject
color_subjects = {}
# to store semester wise subjects
datesheet = [[]]
# create graph with nodes and edges
CompleteGraph = nx.Graph()
x = 1

# for loop in the range of semesters
for i in range(0,s):
    # storing all the subjects
    sub = []
    # complete graph
    G = nx.Graph()
    for j in range(0,len(subjects)):
        if matrix[i][j] == 1:
            # sem wise subject list
            sub.append(reverse[j]) 
    # putting the subjects in the datesheet according to the semesters
    datesheet.append(sub)

    # Code to define a list of colors
    chromatic = [0, 1, 2, 3, 4, 5, 6]
    # The code then iterates through the list of subjects and checks if the subject is 
    # in the color_subjects list. 
    # If it is, then it removes that subject from chromatic.
    # The code removes the color_subjects[reverse[z]] from chromatic.
    for y in range(0, i):
        for z in range(0,len(subjects)):
            if matrix[y][z] == 1 and reverse[z] in sub and color_subjects[reverse[z]] in chromatic:
                chromatic.remove(color_subjects[reverse[z]])    
    index = 0
    # After this, it adds every subject back into color_subjects so that they can be sorted properly. 
    # The code starts by creating a list of subjects
    # It will then update all instances where this new subject has been found with the current chromatic 
    # value for this iteration.
    for subject in range(0, len(sub)):
        if sub[subject] not in color_subjects.keys():
            color_subjects.update({sub[subject]:chromatic[index]})
            index = index + 1
            
    print("Complete Graph for Subject of Sem ", (x)%4, ":")
    print(sub)
    
    x += 1
    G.add_nodes_from(sub)
    G.add_edges_from(itertools.combinations(sub, 2))
    values = [color_subjects.get(node, 0.25) for node in G.nodes()]
    CompleteGraph.add_nodes_from(sub)
    CompleteGraph.add_edges_from(itertools.combinations(sub, 2), weight = 8)
    nx.draw(G, cmap = plt.get_cmap('jet'), node_size = 1500, node_color = values, with_labels = True, font_color = 'white')
    plt.show()
```

```python
x = 1

# for loop starting from semesters to semesters * courses
for i in range(s, n*s):
    # storing all the subjects
    sub = []
    # complete graph
    G = nx.Graph()
    for j in range(0, len(subjects)):
        if matrix[i][j] == 1:
            # sem wise subject list
            sub.append(reverse[j]) 
    # putting the subjects in the datesheet according to the semesters
    datesheet.append(sub)
    
    # Code to define a list of colors
    chromatic = [0, 1, 2, 3, 4, 5, 6]
 
    # The code then iterates through the list of subjects and checks if the subject is 
    # in the color_subjects list. 
    # If it is, then it removes that subject from chromatic.
    # The code removes the color_subjects[reverse[z]] from chromatic.
    for y in range(0, i):
        for z in range(0,len(subjects)):
            if matrix[y][z] == 1 and reverse[z] in sub and color_subjects[reverse[z]] in chromatic:
                chromatic.remove(color_subjects[reverse[z]])
   
    for y in range(i+1, s*n):
        for z in range(0,len(subjects)):
            if matrix[y][z] == 1 and reverse[z] in color_subjects.keys() and color_subjects[reverse[z]] in chromatic:
                chromatic.remove(color_subjects[reverse[z]]) 
                
    index = 0
    # After this, it adds every subject back into color_subjects so that they can be sorted properly. 
    # The code starts by creating a list of subjects
    # It will then update all instances where this new subject has been found with the current chromatic 
    # value for this iteration.
    for subject in range(0, len(sub)):
        if sub[subject] not in color_subjects.keys():
            color_subjects.update({sub[subject]:chromatic[index]})
            index = index + 1
            
    print("Complete Graph for Subject of Sem ", (x)%4, ":")
    print(sub)
    
    x += 1
    G.add_nodes_from(sub)
    G.add_edges_from(itertools.combinations(sub, 2))
    CompleteGraph.add_nodes_from(sub)
    CompleteGraph.add_edges_from(itertools.combinations(sub, 2), weight = 8)
    values = [color_subjects.get(node, 0.25) for node in G.nodes()]
    nx.draw(G, cmap = plt.get_cmap('jet'), node_size = 1500, node_color = values, with_labels = True, font_color = 'white')
    plt.show()
```

```python
# The code starts by creating an empty array of 6x9. This is the size of the data frame that will be created.
data = np.empty((6, 9), dtype = object)

# The code then iterates through each subject, and if a subject has 1 in its corresponding slot, it adds that value 
# to the end of the data frame at index 2+color_subjects[reverse[j]]
# The final result is a DataFrame with all subjects colored according to their course.
# n*s = courses * semesters
for i in range(0, n*s):
    for j in range(0,len(subjects)):
        if matrix[i][j] == 1:
            data[i][2 + color_subjects[reverse[j]]] = reverse[j]
final = pd.DataFrame(data, columns = ['Course', 'Sem', 'Slot1', 'Slot2', 'Slot3', 'Slot4', 'Slot5', 'Slot6', 'Slot7'])

# Here we are creating the course and the semester column in our final schedule
for i in range(0, n*s):
    if i < s:
        final.at[i, 'Course'] = (df['Course'].unique()[0])
    else:
        final.at[i, 'Course'] = (df['Course'].unique()[1])
        
    final.at[i, 'Sem'] = (df['Sem'].unique()[(i)%s])

print("\nFinal Schedule For Each Semester:")
final
```
